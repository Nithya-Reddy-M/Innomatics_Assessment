```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def control(self, action):
        if action.upper() == "ON":
            self.status = "ON"
        elif action.upper() == "OFF":
            self.status = "OFF"
        print(f"{self.name} is {self.status}")


name = input("Light Name: ")
action = input("Action: ")

light = SmartLight(name)
light.control(action)
```

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")

emp = Employee(name, emp_id, department)
emp.display()
```

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


name = input("Contact Name: ")
phone = input("Phone Number: ")

contact = Contact(name, phone)
contact.display()
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


name = input("Product Name: ")
price = float(input("Price: "))

product = Product(name, price)
product.display()
```

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


name = input("Movie Name: ")
rating = float(input("Rating: "))

movie = Movie(name, rating)
movie.display()
```
