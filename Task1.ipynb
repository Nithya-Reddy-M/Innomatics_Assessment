```python
username=input("Username: ")
password=input("Password: ")
crt_username="admin"
crt_password="1234"

if username==crt_username and password==crt_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
marks=[45,78,90,33,60]
cnt1=0
cnt2=0

for m in marks:
    if m>=50:
        cnt1+=1
    else:
        cnt2+=1
print("Total Students:", len(marks))
print("Passed Students:", cnt1)
print("Failed Students:",cnt2)
```

```python
names=["Alice","bob","CHARLIE"]
cleaned_list=[]

for n in names:
    cleaned_n=n.strip().lower()
    cleaned_list.append(cleaned_n)
print(cleaned_list)
```

```python
messages=["Hi", "Welcome to the platform", "OK"]

for m in messages:
    l=len(m)
    print("Message:",m)
    print("Length:",l)

    if l>10:
        print("Flag: Longer than 10 chars")
    print()
```

```python
logs=["INFO", "ERROR", "WARNING", "ERROR"]
err_cnt=0

for l in logs:
    if l=="ERROR":
        err_cnt+=1
print("Number of ERROR entries: ", err_cnt)
```
