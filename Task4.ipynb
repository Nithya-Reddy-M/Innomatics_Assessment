```python
def recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: "))

        if amount < 50:
            print("Recharge amount must be at least 50.")
        elif amount not in valid_plans:
            print("Invalid plan. Valid plans are:", valid_plans)
        else:
            print("Recharge successful!")
            break
recharge()
```

```python
def inventory_check(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")
products = {}
n = int(input("Enter no.of products: "))
for i in range(n):
    name = input("Enter product name: ")
    stock = int(input("Enter stock quantity: "))
    products[name] = stock
inventory_check(products)
```

```python
def student_result(marks):
    average = sum(marks) / len(marks)
    print("Average:", average)
    if average >= 50:
        print("Pass")
    else:
        print("Fail")
user_input = input("Enter student marks separated by space: ")
marks = list(map(int, user_input.split()))
student_result(marks)
```

```python
def cab_fare():
    while True:
        distance = float(input("Enter distance (in km): "))
        peak = input("Is it peak hour? (yes/no): ").lower()
        fare = 50 + (12 * distance)
        if peak == "yes":
            fare *= 1.25
        print("Total Fare:", fare)
        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            break
cab_fare()
```

```python
def attendance_check(attendance):
    total_days = len(attendance)
    present_days = attendance.count("P")
    percentage = (present_days / total_days) * 100
    print("Attendance Percentage:", percentage)
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

user_input = input("Enter attendance (P/A) separated by space: ")
attendance = user_input.upper().split()
result = attendance_check(attendance)
print(result)
```

```python
def password_check(password):
    special_chars = "@#$"
    if len(password) < 8:
        print("Password too short")
        return
    has_digit = any(char.isdigit() for char in password)
    has_special = any(char in special_chars for char in password)
    if not has_digit:
        print("Password must contain at least one digit")
    elif not has_special:
        print("Password must contain at least one special character (@, #, $)")
    else:
        print("Strong Password")
user_password = input("Enter your password: ")
password_check(user_password)
```
