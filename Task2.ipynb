```python
sentence = input("Enter sentence: ")
words = sentence.split()
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
emp={
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

high_emp=max(emp,key=emp.get)
high_sal=emp[high_emp]

print(f"Highest Salary: {high_emp} - {high_sal}")
```

```python
nums=[45, 22, 89, 10, 66]
max= nums[0]
min=nums[0]
for num in nums:
    if num>max:
        max=num
    if num<min:
        min=num
print("List:", nums)
print("Max:", max)
print("Min:", min)
```

```python
prices=[450, 1200, 899, 1500, 300]
cnt=0

for price in prices:
    if price>1000:
        cnt+=1

print("Products above 1000:", cnt)
```

```python
attendance = input("Enter attendance (P/A): ").upper()

present_count = attendance.count('P')
percentage = (present_count / len(attendance)) * 100

print("Attendance Percentage:", percentage)
```

```python
nums_input=input("Enter phone numbers separated by space: ")
phn_list=list(map(int, nums_input.split()))
unique_nums=set(phn_list)
print("Unique phone numbers are:", unique_nums)
```

```python
text=input("Enter a string: ")
char_cnt={}
for char in text:
    if char in char_cnt:
        char_cnt[char]+=1
    else:
        char_cnt[char]=1

print(char_cnt)
```

```python
marks_input = input("Enter marks separated by space: ")
marks = list(map(int, marks_input.split()))

total = sum(marks)
count = len(marks)
average = total / count
print("Average marks:", average)
```
