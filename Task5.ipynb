```python
def post_engagement(likes):
    total_likes = 0
    for like in likes:
        total_likes += like

    print("Total Likes:", total_likes)

    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


user_input = input("Enter likes separated by space: ")
likes = list(map(int, user_input.split()))

post_engagement(likes)
```

```python
def medicine_stock(stock):
    print("Medicine Stock:", stock)

    if stock < 10:
        print("Status: Low Stock Alert")
    else:
        print("Status: Stock Sufficient")


stock = int(input("Enter medicine stock: "))
medicine_stock(stock)
```

```python
def rainfall_check(rainfall, required_level):
    total = 0
    for value in rainfall:
        total += value

    average = total / len(rainfall)
    print("Average Rainfall:", int(average))

    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


user_input = input("Enter rainfall values separated by space: ")
rainfall = list(map(int, user_input.split()))
required_level = float(input("Enter required rainfall level: "))

rainfall_check(rainfall, required_level)
```

```python
def check_duplicates(usernames):
    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


user_input = input("Enter usernames separated by space: ")
usernames = user_input.split()

check_duplicates(usernames)
```

```python
def premium_crops(prices, threshold=2000):
    return [p for p in prices if p > threshold]

prices = list(map(int, input("Enter crop prices separated by space: ").split()))
print("Premium Crops:", premium_crops(prices))
```

```python
def app_health_checker(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    return status

errors = int(input("Enter error count: "))
print("Error Count:", errors)
print("System Status:", app_health_checker(errors))
```

```python
def transaction_checker(amount, limit=50000):
    return "Approved" if amount <= limit else "Rejected"

amount = int(input("Enter transaction amount: "))
print("Transaction Amount:", amount)
print("Transaction Status:", transaction_checker(amount))
```

```python
def attendance_eligibility(attendance_list):
    total_classes = len(attendance_list)
    attended = sum(attendance_list)
    percentage = (attended / total_classes) * 100
    eligibility = "Eligible" if percentage >= 75 else "Not Eligible"
    return percentage, eligibility

attendance_list = list(map(int, input("Enter attendance (1 for present, 0 for absent) separated by space: ").split()))
percentage, eligibility = attendance_eligibility(attendance_list)
print("Attendance Percentage:", percentage)
print("Exam Eligibility:", eligibility)
```

```python
def electricity_bill(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = 100*3 + (units-100)*5
    else:
        bill = 100*3 + 100*5 + (units-200)*7
    
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    
    return bill, usage

units = int(input("Enter electricity units consumed: "))
bill, usage_status = electricity_bill(units)
print("Total Bill:", bill)
print("Usage Status:", usage_status)
```
