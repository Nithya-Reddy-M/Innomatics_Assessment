```python
emp= {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
high_score=max(emp.values())
top_performers=[name for name, score in emp.items() if score == high_score]
print(f"Top Performers Eligible for Bonus: {','.join(top_performers)} (Score: {high_score})")
```

```python
import string
from collections import Counter
query="buy mobile phones, buy phones online"
query=query.lower()
query=query.translate(str.maketrans('', '', string.punctuation))
words=query.split()
word_cnt=Counter(words)
res={word: cnt for word, cnt in word_cnt.items() if cnt>1}
print(res)
```

```python
user_input=input("Sensor_readings:")
sensor_readings=list(map(int, user_input.split()))
valid_read=[]
for hour, reading in enumerate(sensor_readings):
    if reading%2==0:
        valid_read.append((hour, reading))
print("Valid Sensor Readings(Hour, Value): ", valid_read)
```

```python
user_input=input("emails")
emails=[email.strip() for email in user_input.split(",")]
dom_cnt={}
for email in emails:
    dom=email.split("@")[1]
    if dom in dom_cnt:
        dom_cnt[dom]+=1
    else:
        dom_cnt[dom]=1
total_emails=len(emails)

for dom, cnt in dom_cnt.items():
    percentage=(cnt/total_emails)*100
    print(f"{dom}: {percentage:.0f}%")
```

```python
user_input = input("sales: ")
sales = list(map(int, user_input.split()))
average = sum(sales) / len(sales)
threshold = average * 1.30

for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

```python
user_input = input("user_ids: ")
user_ids = [uid.strip() for uid in user_input.split(",")]
cnt_dict = {}

for uid in user_ids:
    if uid in cnt_dict:
        cnt_dict[uid] += 1
    else:
        cnt_dict[uid] = 1

for uid, cnt in cnt_dict.items():
    if cnt > 1:
        print(f"{uid} → {cnt} times")
```
