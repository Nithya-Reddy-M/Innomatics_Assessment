```python
def smart_parking(capacity, logs):
    current = 0
    peak = 0

    for log in logs:
        if log.upper() == "IN":
            current += 1
        elif log.upper() == "OUT" and current > 0:
            current -= 1

        if current > peak:
            peak = current

    print("Currently Parked Vehicles:", current)

    if current > capacity:
        print("Parking Status: Capacity Exceeded")
    else:
        print("Parking Status: Available")


capacity = int(input("Enter parking capacity: "))
user_input = input("Enter vehicle logs separated by space (IN/OUT): ")
logs = user_input.split()

smart_parking(capacity, logs)
```

```python
def estimate_delivery_time(distance, traffic, weather):
    base_time = distance * 5

    if traffic.lower() == "high":
        base_time += 15
    elif traffic.lower() == "medium":
        base_time += 10
    elif traffic.lower() == "low":
        base_time += 5

    if weather.lower() == "rainy":
        base_time += 10
    elif weather.lower() == "stormy":
        base_time += 20
    elif weather.lower() == "clear":
        base_time += 0

    print("Estimated Delivery Time:", base_time, "minutes")


distance = float(input("Distance (km): "))
traffic = input("Traffic Level: ")
weather = input("Weather: ")

estimate_delivery_time(distance, traffic, weather)
```

```python
def seat_analysis(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", str(int(occupancy)) + "%")

    if occupancy == 100:
        print("Show Status: Housefull")
        print("Suggestion: Open Additional Show")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


total_seats = int(input("Total Seats: "))
n = int(input("Number of booked entries: "))

booked_seats = []
for i in range(n):
    booked_seats.append(int(input()))

seat_analysis(total_seats, booked_seats)
```

```python
def server_load(cpu_readings):
    total = 0
    for value in cpu_readings:
        total += value

    average = total / len(cpu_readings)

    print("Average CPU Load:", str(int(average)) + "%")

    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average < 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


n = int(input("Number of CPU readings: "))
cpu_readings = []

for i in range(n):
    cpu_readings.append(float(input()))

server_load(cpu_readings)
```

```python
def resource_tracker(usage):
    overused = []

    for resource, hours in usage.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


n = int(input("Number of resources: "))
usage = {}

for i in range(n):
    name = input("Resource Name: ")
    hours = float(input("Usage Hours: "))
    usage[name] = hours

resource_tracker(usage)
```

```python
def event_registration(capacity, registrations):
    confirmed = 0
    waitlisted = 0

    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    if waitlisted > 0:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


capacity = int(input("Event Capacity: "))
registrations = int(input("Total Registrations: "))

event_registration(capacity, registrations)
```
